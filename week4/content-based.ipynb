```python
features = data[data.columns[0:10]]
```

```python
user1_prefer = (features.T * data['User 1']).sum(axis = 1)
```

```python
user1_prefer
```

```python
user2_prefer = (features.T * data['User 2']).sum(axis = 1)
```

```python
user2_prefer
```

```python
user1_doc = (features * user1_prefer).sum(axis = 1)
```

```python
user1_doc.sort_values(ascending=False)
```

```python
user2_doc = (features * user2_prefer).sum(axis = 1)
```

```python
user2_doc.sort_values(ascending=False)
```

```python
norm_features
```

```python
norm_user1_prefer = (norm_features * data['User 1']).sum(axis  =1)
```

```python
norm_user1_prefer
```

```python
norm_user2_prefer = (norm_features * data['User 2']).sum(axis  =1)
```

```python
norm_user2_prefer
```

```python
norm_user1_doc = (norm_features.T * norm_user1_prefer).sum(axis = 1)
```

```python
norm_user1_doc.sort_values(ascending=False)
```

```python
norm_user2_doc = (norm_features.T * norm_user2_prefer).sum(axis = 1)
```
