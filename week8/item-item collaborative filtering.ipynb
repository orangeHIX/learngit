```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('Assignment 5.xlsx', 'Ratings', index_col=0)
```

```python
data
```

```python
r_data = data[data.columns[0:-1]][0:-1]
```

```python
r_data
```

```python
s = np.sqrt(r_data.applymap(lambda x: x**2).sum())
```

```python
X_square = r_data.fillna(0).T.dot(np.array(r_data.fillna(0)))
```

```python
X_square.columns = X_square.index
```

```python
X_d = np.reshape(s, (s.shape[0], 1)).dot(np.reshape(s, (1, s.shape[0])))
```

```python
sim = (X_square/X_d)
```

```python
sim
```

```python
r_data_norm = (r_data.T - r_data.mean(axis =1)).fillna(0)
```

```python
r_data_norm
```

```python
X_square_norm = r_data_norm.fillna(0).dot(r_data_norm.fillna(0).T)
```

```python
X_square_norm
```

```python
s_norm = np.sqrt(r_data_norm.T.applymap(lambda x: x**2).sum())
```

```python
X_d_norm = np.reshape(s_norm, (s_norm.shape[0], 1)).dot(np.reshape(s_norm, (1, s_norm.shape[0])))
```

```python
sim_norm = X_square_norm/X_d_norm
```
