```python
def top3Prediction(data, data_corr, user_id):
    neighbors =data_corr[user_id].sort_values(ascending = False)[1:6]
    pre =  (data[neighbors.index] * neighbors).sum(axis = 1)/((pd.isnull(data[neighbors.index]).applymap(lambda x: 0 if x else 1)
    *np.tile(neighbors,(data.shape[0],1))).sum(axis =1))
    return pre.sort_values(ascending = False)[:3]
```

```python
top3Prediction(data, data_corr, 3525)
```

```python
top3Prediction(data, data_corr, 3867)
```

```python
top3Prediction(data, data_corr, 89)
```

```python
def top3PredictionNorm(data, data_corr, user_id, user_mean):
    neighbors =data_corr[user_id].sort_values(ascending = False)[1:6]
    pre = user_mean[user_id] + ((data[neighbors.index]-user_mean[neighbors.index])* neighbors).sum(axis = 1)/((pd.isnull(data[neighbors.index]).applymap(lambda x: 0 if x else 1)
    *np.tile(neighbors,(data.shape[0],1))).sum(axis =1))
    return pre.sort_values(ascending = False)[:3]
```

```python
top3PredictionNorm(data, data_corr, 3712, data.mean())
```

```python
top3PredictionNorm(data, data_corr, 3525, data.mean())
```

```python
top3PredictionNorm(data, data_corr, 3867, data.mean())
```

```python
top3PredictionNorm(data, data_corr, 89, data.mean())
```
